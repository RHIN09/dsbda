```python
import pandas as pd

df=pd.read_csv('dataset_Facebook.csv', delimiter=';')

df.head()
```

```python
df.columns
```

```python
# a. create data subsets from type of posts 
photo_posts=df[df['Type']=='Photo']
status_posts=df[df['Type']=='Status']

print(photo_posts.head())
print(status_posts.head())
```

```python
# b. Merge two subsets
merged_posts=pd.concat([photo_posts,status_posts])

print(merged_posts.head())
```

```python
# c. sort Data on page total likes
sorted_data=df.sort_values(by='Page total likes', ascending=False)
print(sorted_data.head(5))
```

```python
# d. Transposing Data (flip rows and columns)
transposed_data=df.head(4).T #transpose first 4 rows
print(transposed_data)
```

```python
# e. Melting data to long format (convert column into rows)
melted_data=pd.melt(df,id_vars=['Type'],value_vars=['like','share','comment'])

print(melted_data.head(5))

print(melted_data.tail(5))

#Each of the columns 'like', 'share', and 'comment' becomes a row, with their column name in variable, and value in value.
```

```python
# f. casting data to wide format (reverse of melting)

wide_data=melted_data.pivot_table(index=None, columns='variable', values='value', aggfunc='sum').reset_index()
print(wide_data)

#
```
