```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('heart.csv')

print(df.head())

df.columns
```

```python
# a. histogram
plt.figure(figsize=(10,4))
plt.hist(df['age'],bins=10,edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('age')
plt.ylabel('count')
```

```python
#b. Dotplot

plt.figure(figsize=(10,4))
plt.plot(df['age'],df['chol'],'o',color='purple')
plt.title("Dot Plot: Age vs Cholesterol ")
plt.xlabel('age')
plt.ylabel('cholesterol')

```

```python
# c. barplot

plt.figure(figsize=(10,4))
sns.barplot(x='age',y='chol',data=df,palette='Set1')
plt.title('Cholestrol by Age Status')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
```

```python
# d. Line Chart
df_sorted=df.sort_values('age')
plt.figure(figsize=(10,4))
plt.plot(df_sorted['age'],df_sorted['chol'],marker='o')
plt.title('Cholesterol vs Age')
plt.xlabel('Age')
plt.ylabel("Cholesterol")
```

```python
# e. Box plot with histogram and scatterplot

# Create a box plot for cholesterol
plt.figure(figsize=(20, 8))

# Box plot
plt.subplot(1, 3, 1)
sns.boxplot(x=df['age'], color='lightblue')
plt.title("Box Plot of Age")

# Histogram
plt.subplot(1, 3, 2)
plt.hist(df['age'], bins=10, color='orange', edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")

# Scatter plot (age vs cholesterol)
plt.subplot(1, 3, 3)
plt.scatter(df['age'], df['chol'], color='purple')
plt.title("Scatter Plot: Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")

plt.show()
```
